```python
import cv2
import os
import sys
import numpy as np

# Add path if needed for local imports
sys.path.append(os.path.abspath("/content/EnlightenGAN-inference"))
from enlighten_inference import EnlightenOnnxModel

def ensure_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)

def enhance_video(input_video_path, output_video_path):
    model = EnlightenOnnxModel()
    ensure_dir(os.path.dirname(output_video_path))

    cap_input = cv2.VideoCapture(input_video_path)

    if not cap_input.isOpened():
        print("Error: Could not open input video.")
        return

    # Get video properties
    frame_width = int(cap_input.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap_input.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap_input.get(cv2.CAP_PROP_FPS)

    # Set up output video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Use 'XVID' or 'avc1' if needed
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    frame_count = 0

    while True:
        ret_input, frame_input = cap_input.read()
        if not ret_input:
            break

        enhanced = model.predict(frame_input)
        out.write(enhanced)
        frame_count += 1

    cap_input.release()
    out.release()

    print(f"Video processing completed. {frame_count} frames enhanced and saved to: {output_video_path}")

if __name__ == "__main__":
    input_video = "/content/test/low_light_video.mp4"
    output_video = "/content/results/enhanced_video.mp4"

    enhance_video(input_video, output_video)

```
